```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
valores_compra = list(map(articulos.get, compra))


# A. Una función que calcule el precio total de la compra. #Reduce
total_compra = reduce((lambda x,y: x + y), valores_compra)
print("El total de la compra es", total_compra)

# B. Una función que calcule el precio total de la compra con un descuento del 10% 
desc_total = lambda n: total_compra - (n * total_compra/100)
print("El total de compra con un 10% es de : ",desc_total(10))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€ 
total_descuento = lambda n: n - (n* (0.10)) if total_compra > 100 else n
print("El precio es:", total_descuento(total_compra))

# D. Una función que calcule el precio total aplicando el IVA (21%)
precio_iva =  lambda n: total_compra + (n * total_compra/100)
print("El precio con iva es: ", precio_iva(21))



# E.lista los artículos cuyo precio es superior a 20€ #filter
superior_20 = list(filter(lambda value: value[1]>20, articulos.items()))
print("Los artículos que son superiores a 20 euros son:", superior_20)
```
